# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# TODO

# Funciones para distribución normal
def dnorm(x, mean=0, sd=1):
    """Densidad de probabilidad (PDF) de la normal"""
    return norm.pdf(x, loc=mean, scale=sd)

def pnorm(q, mean=0, sd=1):
    """Función de distribución acumulativa (CDF) de la normal"""
    return norm.cdf(q, loc=mean, scale=sd)

def qnorm(p, mean=0, sd=1):
    """Función cuantil (inversa de la CDF) de la normal"""
    return norm.ppf(p, loc=mean, scale=sd)

def rnorm(n, mean=0, sd=1):
    """Generación de números aleatorios normales"""
    return norm.rvs(loc=mean, scale=sd, size=n)

# Funciones para distribución uniforme
def dunif(x, low=0, high=1):
    """Densidad de probabilidad (PDF) de la uniforme"""
    return uniform.pdf(x, loc=low, scale=high-low)

def punif(q, low=0, high=1):
    """Función de distribución acumulativa (CDF) de la uniforme"""
    return uniform.cdf(q, loc=low, scale=high-low)

def qunif(p, low=0, high=1):
    """Función cuantil (inversa de la CDF) de la uniforme"""
    return uniform.ppf(p, loc=low, scale=high-low)

def runif(n, low=0, high=1):
    """Generación de números aleatorios uniformes"""
    return uniform.rvs(loc=low, scale=high-low, size=n)

# Ejemplos de uso con distribución normal N(0,1)
np.random.seed(42)
print("=== Distribución Normal N(0,1) ===")
print(f"PDF en x=0: {dnorm(0):.4f}")  # Densidad en 0
print(f"CDF en x=0: {pnorm(0):.4f}")  # Probabilidad P(X ≤ 0)
print(f"Cuantil para p=0.9: {qnorm(0.9):.4f}")  # Valor donde P(X ≤ x) = 0.9
print(f"Muestra de 5 valores: {rnorm(5)}")  # 5 números aleatorios

# Ejemplos de uso con distribución uniforme U(0,1)
print("\n=== Distribución Uniforme U(0,1) ===")
print(f"PDF en x=0.5: {dunif(0.5):.4f}")  # Densidad en 0.5 (siempre 1 para U(0,1))
print(f"CDF en x=0.5: {punif(0.5):.4f}")  # Probabilidad P(X ≤ 0.5)
print(f"Cuantil para p=0.9: {qunif(0.9):.4f}")  # Valor donde P(X ≤ x) = 0.9
print(f"Muestra de 5 valores: {runif(5)}")  # 5 números aleatorios
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Configuración inicial
np.random.seed(42)
mean = 0      # Media (μ)
sd = 1        # Desviación estándar (σ)

# Generar datos aleatorios
data = norm.rvs(loc=mean, scale=sd, size=2000)

# Calcular funciones de distribución
pdf = norm.pdf(data, loc=mean, scale=sd)  # Función de densidad
cdf = norm.cdf(data, loc=mean, scale=sd)  # Función acumulativa
ppf = norm.ppf(np.linspace(0.01, 0.99, 2000), loc=mean, scale=sd)  # Percentiles

# Crear gráficos
fig, axis = plt.subplots(3, 1, figsize=(5, 7))

# PDF (Densidad)
axis[0].plot(sorted(data), norm.pdf(sorted(data), mean, sd), "r-")
axis[0].set_title("Función de Densidad (PDF)")

# CDF (Acumulativa)
axis[1].plot(sorted(data), norm.cdf(sorted(data), mean, sd), "b-")
axis[1].set_title("Función Acumulativa (CDF)")

# PPF (Inversa de la CDF)
axis[2].plot(np.linspace(0.01, 0.99, 2000), ppf, "g-")
axis[2].set_title("Función Percentil (PPF)")

plt.tight_layout()
plt.show()
```
